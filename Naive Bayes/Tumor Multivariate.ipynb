# Multivariate Naive Bayes Classifier 


####  Taking Two Attributes Only

```python
import pandas as pd
from math import e
import numpy as np
import scipy.stats as s
import random
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("data.csv")
dataset.head()
```

```python
data  =  dataset.iloc[0:,1:4]
data
```

```python
Bdata = data[data['diagnosis'] == 'B']
Mdata = data[data['diagnosis'] == 'M']

BTrainData = Bdata[:200]
MTrainData = Mdata[:200]
```

```python
PCapB = 200/400
PCapB
```

```python
PCapBComplement =  1 - PCapB
PCapBComplement
```

```python
CovarianceMatB = BTrainData.cov()
CovarianceMatB = np.array(CovarianceMatB)
CovarianceMatB
```

```python
CovarianceMatM = MTrainData.cov()
CovarianceMatM = np.array(CovarianceMatM)
CovarianceMatM
```

```python
MuCapB = BTrainData.mean()
MuCapB = np.array(MuCapB)
MuCapB.reshape(2,1)
```

```python
MuCapM = np.array(MTrainData.mean())
MuCapM.reshape(2,1)
```

```python
BTestData = Bdata[200:]
MTestData = Mdata[200:]

PBebignTumorB = s.multivariate_normal.pdf(BTestData.iloc[0:,1:3],MuCapB,CovarianceMatB,allow_singular = 1)
PMalignantTumorB = s.multivariate_normal.pdf(BTestData.iloc[0:,1:3],MuCapM,CovarianceMatM,allow_singular = 1)

PBebignTumorM = s.multivariate_normal.pdf(MTestData.iloc[0:,1:3],cov=CovarianceMatB,mean= MuCapB,allow_singular = 1)
PMalignantTumorM = s.multivariate_normal.pdf(MTestData.iloc[0:,1:3],cov=CovarianceMatM,mean= MuCapM,allow_singular = 1)
```

```python
num1 = PBebignTumorB * PCapB
num2 = PMalignantTumorB * PCapBComplement

num3 = PBebignTumorM * PCapB
num4 = PMalignantTumorM * PCapBComplement
```

```python
PredictBTest = num1/(num1+num2)

truenegative = 0
falsepositive = 0
for i in PredictBTest:
    if i >= 0.5:
        truenegative += 1
    else:
        falsepositive += 1
        
PredictedMTest = num3/(num3+num4)

truepositive = 0
falsenegative = 0

for i in PredictedMTest:
    if i >= 0.5:
        truepositive += 1
    else:
        falsenegative += 1
                
print("truenegative: ", truenegative, "falsepositive: ",falsepositive, "truepositive: ", truepositive, "falsenegative: ", falsenegative)
print("Total Benign Tumor: ", len(BTestData))
print("Total Malignant Tumor: ",len(MTestData))

accuracy = (truepositive + truenegative)/(len(BTestData)+len(MTestData))
print("Accuracy: ",accuracy)
```

# Taking all the attributes

```python
CData = dataset.iloc[0:,1:32]
# print(CData)
CBData = CData[CData['diagnosis'] == 'B']
CMData = CData[CData['diagnosis'] == 'M']

CBTrainData = CBData[0:200]
CMTrainData = CMData[0:200]

CBTestData = CBData[200:]
CMTestData = CMData[200:]

TrainData = CBTrainData.append(CMTrainData)
TestData = CBTestData.append(CMTestData)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
classifier = MultinomialNB()
classifier.fit(TrainData.iloc[0:,1:],TrainData['diagnosis'])
```

```python
predictions = classifier.predict(TestData.iloc[0:,1:31])
predictions = predictions == TestData['diagnosis']
t = 0
for i in predictions:
    if i == True:
        t += 1
print(t)
```

```python
predictions = classifier.predict(TestData.iloc[0:,1:31])

truepositiveCount = 0
truenegativeCount = 0

for i in range(1,len(predictions)):
    if predictions[i] == TestData.iloc[i,0]:
        if i == 'B':
            truepositiveCount += 1
        else:
            truenegativeCount += 1
            
print(truepositiveCount,truenegativeCount)


        
```
